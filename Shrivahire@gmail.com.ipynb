During this festive season, various industries witness a surge in sales as people indulge in shopping for a wide range of products, including electronics, apparel, home appliances, and more.

<b>c. Customer Behavior Insights:</b> Analyze customer buying behavior during Diwali, including purchase frequency, average transaction values, and preferred product categories. Identify loyal customers and explore strategies to enhance customer retention.

<b>d. Pricing and Discount Strategies:</b> Investigate the impact of pricing and discount strategies on consumer buying decisions. Determine the most effective pricing ranges and discount levels that drive sales without compromising profitability.

<b>e. Regional Market Variation:</b> Examine sales data from different regions to identify regional variations in consumer preferences and spending habits during Diwali. Tailor marketing and promotional efforts to suit diverse market demands.

# **6. Exploratory Data Analysis**
---

<b>Gender Distribution Analysis</b>

```python
ax=sns.countplot(x='Gender',data=DSD)

for bar in ax.containers:
    ax.bar_label(bar)
```

<b>Distribution of the 'Amount' variable based on gender<b>

```python
paper = plt.figure(figsize = [5,5])
sns.barplot(x='Gender',y='Amount',data=DSD.groupby(
    ['Gender'],as_index=False
)['Amount'].sum().sort_values(by='Amount',ascending=False))
plt.show()
```

From above graphy we can see that Women are the biggest buyer as compare to male

<b>Sales Analysis based on Age category and gender</b>

```python
ax=sns.countplot(x='Age Group',data=DSD,hue='Gender')

for bar in ax.containers:
    ax.bar_label(bar)
```

<b>Sales analysis based on age group</b>

```python
sale_age=DSD.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

ax=sns.barplot(x='Age Group',y='Amount',data=sale_age)

for bars in ax.containers:
    ax.bar_label(bars)
```

From above graph it is clear that most of buyer are from age group 26 to 35

<b>Order Analysis in Top 5 states</b>

```python
sales_state=DSD.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(data=sales_state,x='State',y='Orders')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

From above graph we can say that UP and MH puts order more as compare to other states

<b>Sales Analysis on Top 5 States</b>

```python
sales_state=DSD.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(data=sales_state,x='State',y='Amount')

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

From above graph we can say that most of sales happening in UP, MH KA and Delhi.

<b>Marital Status Analysis</b> 

```python
ax=sns.countplot(data=DSD,x='Marital_Status')
sns.set(rc={'figure.figsize':(5,4)})

for bars in ax.containers:
    ax.bar_label(bars)
```

<b>Sales analysis based on Marital status and gender</b>

```python
sales_state=DSD.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')       
plt.show()
```

From above graph it is clear that most of the buyers are Married women and they have high purchasing power

<b>Sales analysis by customers Occupation</b>

```python
sns.set(rc={'figure.figsize':(15,7)})
ax=sns.countplot(data=DSD,x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

<b>Expenditure based on Occupation</b>

```python
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Occupation',y='Amount')
plt.show()
```

```python
sales_occupation=DSD.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_occupation
```

Above grapgh shows IT, Healthcare and Aviation has contributed in most sales

<b>Top 5 categories and orders placed in each category</b>

```python
sale_prodCat=DSD.groupby(['Product_Category'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)

sns.set({'figure.figsize':(10,5)})
sns.barplot(data=sale_prodCat,x='Product_Category',y='Orders')
plt.show()
```

Above graph clarifies that Clothing & Apparel, Food, Electrnics & Gadgets, Footwear & Shoes and Household items are the top 5 categories for highest sale

<b>Products in each product Category section</b>

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=DSD,x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

<b>Sale across each product categories</b>

```python
sale_prodCat=DSD.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)

sns.set({'figure.figsize':(10,5)})
sns.barplot(data=sale_prodCat,x='Product_Category',y='Amount')
plt.show()
```

Above graph shows that Food, Clothing & Appareal, Electronic & Gadgets, Footwear & Shoes and Furniture are the top 5 categories of sale
